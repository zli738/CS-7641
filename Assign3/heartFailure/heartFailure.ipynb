```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, f1_score, ConfusionMatrixDisplay, confusion_matrix,classification_report
from sklearn.model_selection import learning_curve, validation_curve
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import seaborn as sns
from timeit import default_timer as timer
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, pairwise_distances
from sklearn.random_projection import SparseRandomProjection
from scipy.stats import pearsonr, kurtosis
from sklearn.decomposition import PCA,FastICA
from sklearn.preprocessing import OneHotEncoder
```

```python
### load dataset
data = pd.read_csv("./heart_failure_clinical_records.csv")
```

```python
data.info()
```

```python
y = data['DEATH_EVENT']
X = data.drop(['time','DEATH_EVENT'],axis=1)
```

```python
# Scale the data
X_scaled = X
to_scale = ['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium']
X_to_scale = X[to_scale].to_numpy()
scaling = StandardScaler(with_mean=True, with_std=True).fit_transform(X_to_scale)
# X_scaled = scaling.transform(X_to_scale)
X_scaled[to_scale] = scaling
```

```python
# split training and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
# Fit the GMM model
n_range = np.arange(2,30,2)
log_likelihoods = []
bics = []
aics = []
silhouttes = []
for n in n_range:
    gmm = GaussianMixture(n_components=n, random_state=10)
    gmm.fit(X_train)
    
    # Log-likelihood
    log_likelihood = gmm.score(X_train)
    log_likelihoods.append(log_likelihood)
    
    # BIC and AIC
    bic = gmm.bic(X_train)
    aic = gmm.aic(X_train)
    bics.append(bic)
    aics.append(aic)
    # Predict the cluster labels
    labels = gmm.predict(X_train)
    
    # Silhouette Score
    silhouette = silhouette_score(X_train, labels)
    silhouttes.append(silhouette)
```

```python
fig, axs = plt.subplots(2,2)
axs[0,0].plot(n_range,log_likelihoods)
axs[0,0].set( ylabel='Log Likelihood',xlabel='n cluster')

axs[1,0].plot(n_range,bics)
axs[1,0].set( ylabel='BIC',xlabel='n cluster')

axs[0,1].plot(n_range,aics)
axs[0,1].set(ylabel='AIC',xlabel='n cluster')

axs[1,1].plot(n_range,silhouttes)
axs[1,1].set(ylabel='Silhouette Score',xlabel='n cluster')

fig=plt.gcf()
fig.suptitle('Expectation Maximization')
fig.set_size_inches(10,8)

plt.savefig("./plots/gmm.png")
```

```python
inertias = []
db_indices = []
silhouttes_knn = []
for n in n_range:
    kmeans = KMeans(n_clusters=n, random_state=10)
    kmeans.fit(X_train)
    
    # Log-likelihood
    inertia = kmeans.inertia_
    inertias.append(inertia)

    # Predict the cluster labels
    labels = kmeans.predict(X_train)
    
    # Silhouette Score
    silhouette = silhouette_score(X_train, labels)
    silhouttes_knn.append(silhouette)

    db_index = davies_bouldin_score(X_train, labels)
    db_indices.append(db_index)
```

```python
fig, axs = plt.subplots(3,1)
axs[0].plot(n_range,inertias)
axs[0].set( ylabel='inertia',xlabel='n cluster')

axs[1].plot(n_range,db_indices)
axs[1].set( ylabel='Davies-Bouldin Index',xlabel='n cluster')

axs[2].plot(n_range,silhouttes_knn)
axs[2].set(ylabel='Silhouette Score',xlabel='n cluster')

fig=plt.gcf()
fig.suptitle('K Means')
fig.set_size_inches(10,10)
plt.savefig("./plots/kmeans.png")
```

```python
correlations = []
reconstruction_errors = []
n_components_range = np.arange(1,12)
for n in n_components_range:
    srp = SparseRandomProjection(n_components=n, random_state=10)
    
    # Fit and transform the data
    X_reduced = srp.fit_transform(X_train)
    
    # Compute pairwise distances
    original_distances = pairwise_distances(X_train)
    reduced_distances = pairwise_distances(X_reduced)
    
    # Flatten the distance matrices
    original_distances_flat = original_distances.flatten()
    reduced_distances_flat = reduced_distances.flatten()
    
    # Compute Pearson correlation
    correlation, _ = pearsonr(original_distances_flat, reduced_distances_flat)
    correlations.append(correlation)
    
    # Compute reconstruction error
    reconstruction_error = np.mean((original_distances - reduced_distances) ** 2)
    reconstruction_errors.append(reconstruction_error)
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(n_components_range,correlations)
axs[0].set( ylabel='Pearson correlation',xlabel='n components')

axs[1].plot(n_components_range,reconstruction_errors)
axs[1].set( ylabel='reconstruction error',xlabel='n components')

fig=plt.gcf()
fig.suptitle('Randomized Projection')
fig.set_size_inches(8,8)
plt.savefig("./plots/srp.png")
```

```python
explained_variances = []
reconstruction_errors_pca = []
n_components_range = np.arange(1,12)
for n in n_components_range:
    pca = PCA(n_components=n, random_state=10)
    
    # Fit the model to the data
    pca.fit(X_train)
    
    # Explained variance for each principal component
    explained_variance = pca.explained_variance_ratio_
    
    # Total explained variance
    total_explained_variance = np.sum(explained_variance)
    explained_variances.append(total_explained_variance)
    
    X_reduced = pca.transform(X_train)
    
    # Reconstruct the data from the reduced space
    X_reconstructed = pca.inverse_transform(X_reduced)
    
    # Compute the reconstruction error
    reconstruction_error = np.mean((X_train - X_reconstructed) ** 2)
    reconstruction_errors_pca.append(reconstruction_error)
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(n_components_range,explained_variances)
axs[0].set( ylabel='Explained Variance',xlabel='n components')

axs[1].plot(n_components_range,reconstruction_errors_pca)
axs[1].set( ylabel='reconstruction error',xlabel='n components')

fig=plt.gcf()
fig.suptitle('PCA')
fig.set_size_inches(8,8)
plt.savefig("./plots/pca.png")
```

```python
reconstruction_errors_ica = []
kurtosis_ica = []
n_components_range = np.arange(1,12)
for n in n_components_range:
    ica = FastICA(n_components=n, random_state=10)
    
    # Fit the model to the data
    ica.fit(X_train)
    
    X_reduced = ica.transform(X_train)
    kurtosis_value = kurtosis(X_reduced, axis=0).mean()
    kurtosis_ica.append(kurtosis_value)
    
    # Reconstruct the data from the reduced space
    X_reconstructed = ica.inverse_transform(X_reduced)
    
    # Compute the reconstruction error
    reconstruction_error = np.mean((X_train - X_reconstructed) ** 2)
    reconstruction_errors_ica.append(reconstruction_error)
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(n_components_range,kurtosis_ica)
axs[0].set( ylabel='kurtosis',xlabel='n components')

axs[1].plot(n_components_range,reconstruction_errors_ica)
axs[1].set( ylabel='reconstruction error',xlabel='n components')

fig=plt.gcf()
fig.suptitle('ICA')
fig.set_size_inches(8,8)
plt.savefig("./plots/ica.png")

```
