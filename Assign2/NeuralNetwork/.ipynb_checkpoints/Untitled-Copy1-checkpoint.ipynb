```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, ConfusionMatrixDisplay, confusion_matrix
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import learning_curve, validation_curve
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
import seaborn as sns
import mlrose_hiive as mh
from timeit import default_timer as timer
```

```python
### load dataset
data = pd.read_csv("./mushroom_cleaned_data.csv")
```

```python
#### preprocessing data
y = data['class']
X = data.drop('class',axis=1)

# Scale the data
X_scaled = X
X_to_scale = X[['cap-diameter','stem-height','stem-width','season']].to_numpy()
scaling = MinMaxScaler(feature_range=(-1,1)).fit(X_to_scale)
# X_scaled = scaling.transform(X_to_scale)
X_scaled[['cap-diameter','stem-height','stem-width','season']] = scaling.transform(X_to_scale)
```

```python
# split training and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)
```

```python
#back propogation
nn_bp = MLPClassifier(hidden_layer_sizes=(40,40), activation='relu')
```

```python
fit_start = timer()
nn_bp.fit(X_train, y_train)
fit_end = timer()
print("Wall Clock Time for packpropogation fitting: ", fit_end - fit_start)
```

```python
y_pred_nn_bp = nn_bp.predict(X_test)
nn_bp_pred_score = accuracy_score(y_test, y_pred_nn_bp)
print("Predicting score of packpropogation: ", nn_bp_pred_score)
print(classification_report(y_test, y_pred=y_pred_nn_bp))
```

```python
# Simulated Annealing
nn_sa = mh.NeuralNetwork(hidden_nodes=[40], activation='relu',
                           algorithm = 'simulated_annealing',
                           random_state=12345,
                            learning_rate = .1, max_iters=100, 
                            curve = True)
```

```python
fit_start = timer()
nn_sa.fit(X_train, y_train)
fit_end = timer()
print("Wall Clock Time for SA fitting: ", fit_end - fit_start)
```

```python
y_pred_nn_sa = nn_sa.predict(X_test)
nn_sa_pred_score = accuracy_score(y_test, y_pred_nn_sa)
print("Predicting score of SA: ", nn_sa_pred_score)
print(classification_report(y_test, y_pred=y_pred_nn_sa))
```

```python
# Genetic Algo
nn_ga = mh.NeuralNetwork(hidden_nodes=[1], activation='relu',
                           algorithm = 'genetic_alg',
                           early_stopping = False,
                          random_state=12345,
                         pop_size=250, mutation_prob=0.4,
                            learning_rate = .1,
                            curve = True)
```

```python
fit_start = timer()
nn_ga.fit(X_train, y_train)
fit_end = timer()
print("Wall Clock Time for GA fitting: ", fit_end - fit_start)
```

```python
y_pred_nn_ga = nn_ga.predict(X_test)
nn_ga_pred_score = accuracy_score(y_test, y_pred_nn_ga)
print("Predicting score of GA: ", nn_ga_pred_score)
print(classification_report(y_test, y_pred=y_pred_nn_ga))
```

```python
# Random Heal Climbing
nn_rhc = mh.NeuralNetwork(hidden_nodes=[1], activation='relu',
                           algorithm = 'random_hill_climb',
                           early_stopping = False,
                         random_state=12345,
                         restarts=3,
                            learning_rate = .1,
                            curve = True)
```

```python
fit_start = timer()
nn_rhc.fit(X_train, y_train)
fit_end = timer()
print("Wall Clock Time for RHC fitting: ", fit_end - fit_start)
```

```python
# Predict Accuracy
y_pred_nn_rhc = nn_rhc.predict(X_test)
nn_rhc_pred_score = accuracy_score(y_test, y_pred_nn_rhc)
print("Predicting score of RHC: ", nn_rhc_pred_score)
print(classification_report(y_test, y_pred=y_pred_nn_rhc))
```

```python
# Funtion to plot curves
def curve_plot(x, train_scores, test_scores, title, x_label,filepath=None):
    plt.figure()
    plt.plot(x,np.mean(train_scores,axis=1), label='training')
    plt.plot(x,np.mean(test_scores,axis=1), label='validation')
    
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel("Accuracy Score")
    plt.legend()
    if filepath:
        plt.savefig(filepath)
    # plt.show()
```

```python
# Neural Network Learning Curves
train_sizes_bp, train_scores_bp, test_scores_bp, fit_times_bp, _ = learning_curve(nn_bp, X_train, y_train,cv=5,return_times=True)

curve_plot(train_sizes_bp, train_scores_bp, test_scores_bp, "Neural Networks Learning Curve (1)", 'Train Size')
curve_plot(np.mean(fit_times_bp,axis=1), train_scores_bp, test_scores_bp, "Neural Networks Learning Curve (2)", 'Train Time')
```

```python
train_sizes_sa, train_scores_sa, test_scores_sa, fit_times_sa, _ = learning_curve(nn_sa, X_train, y_train,cv=5,return_times=True)

curve_plot(train_sizes_sa, train_scores_sa, test_scores_sa, "Neural Networks Learning Curve (1)", 'Train Size')
curve_plot(np.mean(fit_times_sa,axis=1), train_scores_sa, test_scores_sa, "Neural Networks Learning Curve (2)", 'Train Time')
```

```python
train_sizes_ga, train_scores_ga, test_scores_ga, fit_times_ga, _ = learning_curve(nn_ga, X_train, y_train,cv=5,return_times=True)
```

```python
curve_plot(train_sizes_ga, train_scores_ga, test_scores_ga, "Neural Networks Learning Curve (1)", 'Train Size')
curve_plot(np.mean(fit_times_ga,axis=1), train_scores_ga, test_scores_ga, "Neural Networks Learning Curve (2)", 'Train Time')
```

```python
train_sizes_rhc, train_scores_rhc, test_scores_rhc, fit_times_rhc, _ = learning_curve(nn_rhc, X_train, y_train,cv=5,return_times=True)
```

```python
curve_plot(train_sizes_rhc, train_scores_rhc, test_scores_rhc, "Neural Networks Learning Curve (1)", 'Train Size')
curve_plot(np.mean(fit_times_rhc,axis=1), train_scores_rhc, test_scores_rhc, "Neural Networks Learning Curve (2)", 'Train Time')
```
